```python
from __future__ import division

import pandas as pd
import time
import csv
import numpy as np
import datetime
import re
import warnings
from annoy import AnnoyIndex

warnings.filterwarnings("ignore")
```

```python
behaviors = pd.read_csv('./data/20150701094451-Behavior_training.csv').drop('date_hour',axis=1)
behaviors_wide = pd.pivot_table(behaviors, values=["score"],
                             index=["user_id","video_id"],
                             aggfunc=np.mean).unstack().fillna(0)
```

```python
behaviors_wide[:5]
```

```python
annoy = AnnoyIndex(len(behaviors.video_id.unique()),metric='angular')
```

```python
users = behaviors.user_id.unique()
users.sort()
```

```python
i = 0
for uid in behaviors.user_id.unique():
    annoy.add_item(uid, behaviors_wide.values[i])
    i += 1
```

```python
str(datetime.datetime.now())
```

```python
annoy.build(1000)
```

```python
str(datetime.datetime.now())
```

```python
annoy.save('viki-1000.ann')
```
